## **Exploratory Data Analysis (EDA) And Visualizations** 

```python
# Summary
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df['Date submitted'] = pd.to_datetime(df['Date submitted'])
df['Date received'] = pd.to_datetime(df['Date received'])

df['Year'] = df['Date submitted'].dt.year
df['Month'] = df['Date submitted'].dt.month
df['Resolution_Days'] = (df['Date received'] - df['Date submitted']).dt.days
```

**UNIVARIATE ANALYSIS**

**Complaints by Year**

```python
plt.figure(figsize=(8,4))

df['Year'].value_counts().sort_index().plot(kind='bar')

plt.title("Distribution of Complaints by Year")
plt.xlabel("Year")
plt.ylabel("Number of Complaints")

plt.tight_layout()
plt.show()
```

**Insight:**
Complaints show an increasing distribution across years, indicating growing customer awareness and reporting activity.

**Top 10 Products**

```python
plt.figure(figsize=(10,5))
sns.countplot(
    y='Product',
    data=df,
    order=df['Product'].value_counts().head(10).index
)
plt.title("Top 10 Products by Complaint Volume")
plt.show()
```

**Insights:** Banking-related products dominate the top positions.

**Timely Response Distribution**

```python
plt.figure(figsize=(6,4))
df['Timely response?'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title("Timely Response Distribution")
plt.ylabel("")
plt.show()
```

**Insights** : A large majority of complaints received a timely response.

**BIVARIATE ANALYSIS**

**Product vs Timely Response**

```python
plt.figure(figsize=(10,5))
sns.countplot(
    y='Product',
    hue='Timely response?',
    data=df,
    order=df['Product'].value_counts().head(7).index
)
plt.title("Timely Response by Product")
plt.show()
```

**Insights:** Most products show a high proportion of timely responses

**Monthly Complaint Trend**

```python
monthly = df.groupby(['Year','Month']).size().reset_index(name='Complaints')

plt.figure(figsize=(12,5))
sns.lineplot(data=monthly, x='Month', y='Complaints', hue='Year')
plt.title("Monthly Complaint Trends by Year")
plt.show()
```

**Insights**:Complaint volumes exhibit seasonal patterns, with similar monthly trends recurring across years, suggesting predictable customer behavior during specific periods.

**Top 10 States**

```python
plt.figure(figsize=(10,5))
sns.barplot(
    y=df['State'].value_counts().head(10).index,
    x=df['State'].value_counts().head(10).values
)
plt.title("Top 10 States by Complaints")
plt.show()
```

**Insight**: California is top amoung this

**MULTIVARIATE ANALYSIS**

**Resolution Time by Product**

```python
plt.figure(figsize=(10,5))
sns.boxplot(
    y='Product',
    x='Resolution_Days',
    data=df,
    order=df['Product'].value_counts().head(6).index
)
plt.title("Resolution Time by Product")
plt.show()
```

Insights: For all products, the median resolution time is close to zero.

This means most complaints are resolved very quickly, often within a few days.

**Submission Channel vs Resolution Time**

```python
plt.figure(figsize=(8,5))
sns.boxplot(
    x='Submitted via',
    y='Resolution_Days',
    data=df
)
plt.title("Resolution Time by Submission Channel")
plt.show()
```

**Insights:**: Digital channels enable faster resolution on average

**Company Response vs Timely Response**

```python
plt.figure(figsize=(10,5))
sns.countplot(
    y='Company response to consumer',
    hue='Timely response?',
    data=df
)
plt.title("Company Response vs Timely Response")
plt.show()
```

Insight: The analysis shows that the majority of complaints receive timely responses across all resolution types, with “Closed with explanation” being the most frequent outcome. Delayed responses are minimal, indicating strong service-level compliance and efficient complaint handling processes.

**Operational Efficiency Analysis**

```python
import seaborn as sns
import matplotlib.pyplot as plt

pair_data = df[['Resolution_Days', 'Year', 'Month']]

sns.pairplot(
    pair_data,
    diag_kind='kde',
    corner=True
)

plt.suptitle(
    "Pair Plot: Resolution Time vs Complaint Time Period",
    y=1.02
)
plt.show()
```

**Product & Response Performance Analysis**

```python
plt.figure(figsize=(10,5))

heatmap_data = pd.crosstab(
    df['Product'],
    df['Timely response?']
)

sns.heatmap(
    heatmap_data,
    annot=True,
    fmt='d',
    cmap='YlGnBu'
)

plt.title("Heatmap: Product vs Timely Response")
plt.xlabel("Timely Response")
plt.ylabel("Product")
plt.show()
```

**Product × State × Timely Response**

```python
top_products = df['Product'].value_counts().head(5).index

plt.figure(figsize=(12,6))
sns.countplot(
    data=df[df['Product'].isin(top_products)],
    x='State',
    hue='Timely response?'
)
plt.xticks(rotation=45)
plt.title("Timely Response Across States for Top Products")
plt.show()
```

**Pivot Table 1: Product vs Timely Response**

```python
pivot_product_timely = pd.pivot_table(
    df,
    index='Product',
    columns='Timely response?',
    values='Complaint ID',
    aggfunc='count',
    fill_value=0
)

pivot_product_timely.head()
```

**Pivot Table 2: State vs Average Resolution Days**

```python
pivot_state_resolution = pd.pivot_table(
    df,
    index='State',
    values='Resolution_Days',
    aggfunc='mean'
).sort_values(by='Resolution_Days', ascending=False)

pivot_state_resolution.head(10)
```

**Correlation Heatmap**

```python
## Correlation Code
corr_matrix = df[['Resolution_Days', 'Year', 'Month']].corr()
corr_matrix
```

```python

plt.figure(figsize=(7,5))
sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

### **Key Insights from the Analysis**

**Complaint Volume Trends**

Complaints vary significantly across years, showing fluctuations in consumer concerns.

A few financial products contribute to the majority of complaints, indicating high-risk service areas.

Insight: Companies should closely monitor these high-complaint products.

**Product-Based Analysis**

Products such as credit cards, mortgages, and checking accounts generate the highest number of complaints.

Certain products also exhibit longer resolution times, indicating complexity in handling these issues.

Insight: Product complexity directly impacts resolution efficiency.

**Timely Response Performance**

The majority of complaints receive timely responses.

Delayed responses form only a small portion of total complaints.

Insight: Overall operational efficiency is strong, but delayed cases require attention.

**Company Response Outcomes**

Most complaints are closed with explanation, followed by monetary relief.

Financial compensation cases are generally handled within stipulated timelines.

Insight: Clear communication resolves most complaints without financial impact.

**Resolution Time Analysis**

Resolution time varies significantly by:

Product type

Submission channel

Web and phone submissions dominate complaint volume but show consistent resolution times.

Insight: Digital channels are effective for complaint handling.

**Geographical Distribution**

Certain states contribute disproportionately to complaint volume.

High-complaint states may reflect higher population density or financial activity.

Insight: Region-specific monitoring is necessary.

**Operational Efficiency**

Very few complaints remain “In progress”.

Indicates efficient closure processes and low backlog.

Insight: Complaint lifecycle management is effective.

### **Overall Conclusion**

The analysis demonstrates that financial institutions generally maintain strong complaint resolution practices, with most complaints addressed in a timely and structured manner. However, specific products and regions consistently generate higher complaint volumes and longer resolution times, highlighting areas for process improvement.
